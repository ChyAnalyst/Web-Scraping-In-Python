```python
from bs4  import BeautifulSoup as soup
import requests


pages = [str(i) for i in range(1,3)]
    
#for every page in the interval 1-2
for page in pages:
       
    #make a get request
    html_soup =requests.get ("https://www.jumia.com.ng/mobile-phones/?page="+ page).text

        
    #parse the content of the request with BeautifulSoup
    page_soup = soup (html_soup,"html.parser")

     #select all the phones on a single page
    containers = page_soup.findAll("div",{"class":"info"})

    #for every single of these phones
    for container in containers:
           
        #scrape the name
        Phone_name = container.find("h3",{"class":"name"}).text
        print(Phone_name)
    
         #scrape the price
        Phone_price = container.find("div",{"class":"prc"}).text
        print(Phone_price)
    

 

   
```
